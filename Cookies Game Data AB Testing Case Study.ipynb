```python
# One function for describing data
def load(path, info = True):
    
    import pandas as pd
    import io
    
    if len(path.split(".csv")) > 1:
        read = pd.read_csv(path)
    elif len(path.split(".xlsx")) > 1:
        read = pd.read_excel(path)
    
    if info:
        if len(read) > 0:
            print("# Data imported!")
            print("# ------------------------------------", "\n")
        
            print("# DIMENSIONS -------------------------")
            print("Observation:", read.shape[0], "Column:", read.shape[1], "\n")
    
            print("# DTYPES -----------------------------")
            if len(read.select_dtypes("object").columns) > 0:
                print("Object Variables:", "\n", "# of Variables:", 
                      len(read.select_dtypes("object").columns), "\n", 
                      read.select_dtypes("object").columns.tolist(), "\n")
    
            if len(read.select_dtypes("integer").columns) > 0:
                print("Integer Variables:", "\n", "# of Variables:", 
                      len(read.select_dtypes("integer").columns), "\n", 
                      read.select_dtypes("integer").columns.tolist(), "\n")
    
            if len(read.select_dtypes("float").columns) > 0:
                print("Float Variables:", "\n", "# of Variables:", 
                      len(read.select_dtypes("float").columns), "\n", 
                      read.select_dtypes("float").columns.tolist(), "\n")
    
            if len(read.select_dtypes("bool").columns) > 0:
                print("Bool Variables:", "\n", "# of Variables:", 
                      len(read.select_dtypes("bool").columns), "\n", 
                      read.select_dtypes("bool").columns.tolist(), "\n")
    
            print("# MISSING VALUE ---------------------")
            print("Are there any missing values? \n ", np.where(read.isnull().values.any() == False, 
                                                            "No missing value!", "Data includes missing value!"), "\n")
            
            buf = io.StringIO()
            read.info(buf=buf)
            info = buf.getvalue().split('\n')[-2].split(":")[1].strip()
            print("# MEMORY USAGE ---------------------- \n", info)
          
        else:
            print("# Data did not import!")
    
    return read
```

# 6. A/B Testing

* Assumptions / Varsayımlar
- Check normality / Önce normalliği kontrol edelim. (shapiro)
- If Normal Distribution, check homogeneity / Normal dağılım ise homojenliği kontrol edelim.

* Steps - Adımlar
- Split & Define Control Group & Test Group / Kontrol Grubunu ve Test Grubunu Böl ve Tanımla
- Apply Shapiro Test for normality / Normallik için Shapiro Testini uygulayın (shapiro)
- If parametric apply Levene Test for homogeneity of variances / Parametrik ise varyansların homojenliği için Levene Testi uygulanır (levene)
- If Parametric + homogeneity of variances apply T-Test / Parametrik + varyansların homojenliği T-Testi uygulanırsa
- If Parametric - homogeneity of variances apply Welch Test / Parametrik - varyansların homojenliği Welch Testi uyguluyorsa
- If Non-parametric apply Mann Whitney U Test directly / Parametrik değilse Mann Whitney U Testini doğrudan uygulayın

```python
# Define A/B groups
ab["version"] = np.where(ab.version == "gate_30", "A", "B")
ab.head()
```

```python
# A-B Testing Function - Quick Solution / A-B Testing Fonksiyonu tanımlıyorum
def AB_Test(dataframe, group, target): # test yapacağım dataframe, test grubuna ayrılan sütun, targetımız olan gameround
    
    # Packages / kullanacağımız kütüphaneler
    from scipy.stats import shapiro
    import scipy.stats as stats
    
    # Split A-B / A-B gruplarını oluşturuyoruz.
    groupA = dataframe[dataframe[group] == "A"][target]
    groupB = dataframe[dataframe[group] == "B"][target]
    
    # Assumption: Normality / Her bir grup için varsayımımızın normalliğini shapiro ile kontrolle T-F atıyoruz
    ntA = shapiro(groupA)[1] < 0.05
    ntB = shapiro(groupB)[1] < 0.05
    # H0: Distribution is Normal! - False / Eğer dağılım normalse H0
    # H1: Distribution is not Normal! - True / Eğer dağılım normal değilse H1
    
    if (ntA == False) & (ntB == False): # "H0: Normal Distribution" / Her ikisi 0.05'ten büyük False
        # Parametric Test / Parametrik Test
        # Assumption: Homogeneity of variances / Varyansların homojenliğine levene testi ile bakıyoruz.
        leveneTest = stats.levene(groupA, groupB)[1] < 0.05
        # H0: Homogeneity: False 
        # H1: Heterogeneous: True 
        
        if leveneTest == False: # / Burada False H0 durumunu grupA-B levene testinde de False koşulu 
            # Homogeneity # Homojenlikte t_test indikatörüne bakıyoruz.
            ttest = stats.ttest_ind(groupA, groupB, equal_var=True)[1]
            # H0: M1 == M2 - False
            # H1: M1 != M2 - True
        else:
            # Heterogeneous / Burada True H1 durumunu grupA-B levene testinde de True koşulu
            ttest = stats.ttest_ind(groupA, groupB, equal_var=False)[1]
            # H0: M1 == M2 - False
            # H1: M1 != M2 - True
    else:
        # Non-Parametric Test / Parametrik olmayan Test
        ttest = stats.mannwhitneyu(groupA, groupB)[1] 
        # H0: M1 == M2 - False
        # H1: M1 != M2 - True
        
    # Result
    temp = pd.DataFrame({
        "AB Hypothesis":[ttest < 0.05], 
        "p-value":[ttest]
    })
    temp["Test Type"] = np.where((ntA == False) & (ntB == False), "Parametric", "Non-Parametric")
    temp["AB Hypothesis"] = np.where(temp["AB Hypothesis"] == False, "Fail to Reject H0", "Reject H0")
    temp["Comment"] = np.where(temp["AB Hypothesis"] == "Fail to Reject H0", "A/B groups are similar!", "A/B groups are not similar!")
    
    # Columns
    if (ntA == False) & (ntB == False):
        temp["Homogeneity"] = np.where(leveneTest == False, "Yes", "No")
        temp = temp[["Test Type", "Homogeneity","AB Hypothesis", "p-value", "Comment"]]
    else:
        temp = temp[["Test Type","AB Hypothesis", "p-value", "Comment"]]
    
    # Print Hypothesis
    print("# A/B Testing Hypothesis")
    print("H0: A == B")
    print("H1: A != B", "\n")
    
    return temp
    
    
    
# Apply A/B Testing / ab dataframemimizde versiyonları sum_gamerounds targetımıza göre AB_testimizi uygulayalım.
AB_Test(dataframe=ab, group = "version", target = "sum_gamerounds")
```
